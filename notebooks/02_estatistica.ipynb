```python
import pandas as pd
from scipy.stats import chi2_contingency
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/retail_sales_dataset.csv')
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df['Month'] = df.index.month_name()
df['Month_Num'] = df.index.month
df.head()
```

```python
# Correlação Pearson
corr_pearson = df['Age'].corr(df['Total Amount'], method='pearson')


print("Correlação Pearson:", corr_pearson)
```

A relação idade e total gasto não tem veracidade visto que a correlação é muito próxima de 0

```python
corr_qty_price = df['Quantity'].corr(df['Price per Unit'], method='pearson')
print("Correlação Quantidade x Preço:", corr_qty_price)
```

A relação quantidade e total gasto não tem veracidade visto que a correlação é muito próxima de 0

```python
import scipy.stats as stats

# Ticket médio por gênero
male_sales = df[df['Gender'] == 'Male']['Total Amount']
female_sales = df[df['Gender'] == 'Female']['Total Amount']

# Teste t de Student (assumindo variâncias diferentes)
t_stat, p_value = stats.ttest_ind(male_sales, female_sales, equal_var=False)

print("t-stat:", t_stat)
print("p-value:", p_value)

# Interpretação
if p_value < 0.05:
    print("Diferença significativa (rejeita H0).")
else:
    print("Diferença não significativa (não rejeita H0).")
```

```python
# 1) Criar tabela de contingência (frequências observadas)
tabela = pd.crosstab(df["Gender"], df["Product Category"])
print("Tabela de contingência:")
print(tabela)

# 2) Aplicar teste qui-quadrado
qui2, p_valor, graus_lib, freq_esperadas = chi2_contingency(tabela)

print("\nResultado do teste:")
print(f"Qui-quadrado: {qui2:.4f}")
print(f"P-valor: {p_valor:.4f}")
print(f"Graus de liberdade: {graus_lib}")
print("Frequências esperadas:")
print(freq_esperadas)
```

Não há evidência estatística de que gênero influencie a categoria de produto comprada.
